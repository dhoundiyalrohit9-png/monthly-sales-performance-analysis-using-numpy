```python
#import numpy
#all sales value are in rupees
import numpy as np
month_names=["January","February","March","April","May","June","July","August","September","October","November","December"]
sale_list=[]
for month_name in month_names:
  sale=float(input(f"enter the sales for {month_name}: "))
  sale_list.append(sale)
sale_array=np.array(sale_list)
total_sales=np.sum(sale_array)
print("total monthly sales=",total_sales)

#Adds all monthly sales values

average_sale=np.mean(sale_array)
print("average monthly sales=",average_sale)
print("Maximum Sales =", np.max(sale_array))
print("Minimum Sales =", np.min(sale_array))

highest_month_index = np.argmax(sale_array)
highest_month = month_names[highest_month_index]


print("Highest sales on =", highest_month)
#in busy months
lowest_month_index = np.argmin(sale_array)
lowest_month = month_names[lowest_month_index]
print("lowest sales on =",lowest_month)
#month with lowest sales

first_half = sale_array[:6]
second_half = sale_array[6:]

print("First Half Sales =", np.sum(first_half))
print("Second Half Sales =", np.sum(second_half))
```
